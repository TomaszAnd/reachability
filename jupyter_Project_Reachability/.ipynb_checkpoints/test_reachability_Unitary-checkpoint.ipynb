```python
import numpy as np
import matplotlib.pyplot as plt

# Question:
# Is it possible to reach any SU(n) given only p randomly sampled generators out of the n^2-1 generators
# or given a vector and p randomly rotated vectors. How probable is it that the vector and the p rotated vectors
# are linearly independen? Very likely!

n = 3
p = 3
Nsample = 300

def random_rot(n):
    M = np.random.randn(n, n)
    Orand, R = np.linalg.qr(M)
    return Orand

# Reference vector
v0 = np.zeros(n)
v0[0]=1

xi_mat = np.zeros((n,p))
for i in range(p):
    Orot = random_rot(n)
    xi_mat[:,i] = Orot @ v0

# Rank criterion
np.linalg.matrix_rank(xi_mat)  == np.linalg.matrix_rank(np.concatenate([xi_mat, v0.reshape(n,1)], axis=1))


# Answere: 
# Hitting the hyperplane spanned by p basis vectors with the v0 is very unlikely!
    
```
