```python
import pickle
import matplotlib.pyplot as plt
import numpy as np

Dlist = [20, 30, 40, 50]
Nsample = 1000

fig, ax = plt.subplots(figsize=(8, 5))

for i, Dval in enumerate(Dlist):
    datastr = "data_D="+str(Dval)+"_Nsample=1000"
    data = pickle.load(open("data_reach/"+datastr,"rb"))
    D = data[2]
    nfinal = data[3]
    Wfinal = data[4]
    x = nfinal/Dval**2
    y = np.sum(Wfinal,axis=1)/Nsample
    ax.semilogy(x,y,"o--",label = r"D = "+str(D))
    ax.set_xlabel(r"$\frac{n}{D^2}$",fontsize=18)
    ax.set_ylabel(r"$\mathcal{P}_{\mathrm{reachable}}$",fontsize=18)
    
ax.legend()
    
    
    
```
